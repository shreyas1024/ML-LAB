```python
# %% [markdown]
# # Credit Card Customer Segmentation using Clustering
#
# This notebook demonstrates customer segmentation using 4 different clustering methods on a credit card dataset.

# %%
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, SpectralClustering
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import warnings
warnings.filterwarnings('ignore')

# %%
# Set random seed for reproducibility
np.random.seed(42)

# %% [markdown]
# ## 1. Generate Sample Credit Card Dataset
#
# If you already have a dataset, you can skip this step and load your own CSV file.

# %%
# Generate synthetic credit card customer data
num_customers = 1000

data = {
    'CustomerID': np.arange(1, num_customers+1),
    'Age': np.random.randint(18, 70, size=num_customers),
    'Gender': np.random.choice(['Male', 'Female'], size=num_customers),
    'Annual_Income': np.random.normal(50000, 15000, num_customers).astype(int),
    'Credit_Limit': np.random.normal(10000, 3000, num_customers).astype(int),
    'Total_Transactions': np.random.poisson(50, num_customers),
    'Avg_Transaction_Value': np.random.normal(150, 50, num_customers).astype(int),
    'Balance': np.random.normal(3000, 1000, num_customers).astype(int),
    'Delinquency': np.random.binomial(1, 0.1, num_customers),
    'Card_Category': np.random.choice(['Blue', 'Silver', 'Gold', 'Platinum'],
                                    size=num_customers, p=[0.6, 0.25, 0.1, 0.05])
}

# Create DataFrame
df = pd.DataFrame(data)

# Add some correlation between features
df['Credit_Limit'] = df['Credit_Limit'] + df['Annual_Income']/10
df['Avg_Transaction_Value'] = df['Avg_Transaction_Value'] + df['Annual_Income']/1000
df['Balance'] = df['Balance'] + df['Credit_Limit']/5

# Ensure no negative values
df['Annual_Income'] = df['Annual_Income'].abs()
df['Credit_Limit'] = df['Credit_Limit'].abs()
df['Avg_Transaction_Value'] = df['Avg_Transaction_Value'].abs()
df['Balance'] = df['Balance'].abs()

# Save to CSV
df.to_csv('credit_card_customers.csv', index=False)

# %%
# Display the first few rows of the dataset
df.head()

# %% [markdown]
# ## 2. Data Preprocessing

# %%
# Load the dataset (using the one we just created)
df = pd.read_csv('credit_card_customers.csv')

# %%
# Basic info about the dataset
print("Dataset Info:")
print(df.info())

# %%
# Descriptive statistics
print("\nDescriptive Statistics:")
df.describe()

# %%
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# %%
# Convert categorical variables to numerical
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
df['Card_Category'] = df['Card_Category'].map({'Blue': 0, 'Silver': 1, 'Gold': 2, 'Platinum': 3})

# Select features for clustering
features = ['Age', 'Annual_Income', 'Credit_Limit', 'Total_Transactions',
            'Avg_Transaction_Value', 'Balance', 'Delinquency', 'Card_Category']
X = df[features]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# %% [markdown]
# ## 3. Dimensionality Reduction for Visualization

# %%
# Apply PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Apply t-SNE for visualization (alternative to PCA)
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# %%
# Plot the reduced data
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.5)
plt.title('PCA Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], alpha=0.5)
plt.title('t-SNE Visualization')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')

plt.tight_layout()
plt.show()

# %% [markdown]
# ## 4. Determine Optimal Number of Clusters (Elbow Method)

# %%
# Elbow method to find optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()

# %% [markdown]
# ## 5. Apply 4 Clustering Methods

# %%
# Function to evaluate clustering
def evaluate_clustering(X, labels):
    if len(np.unique(labels)) > 1:  # Silhouette score requires at least 2 clusters
        score = silhouette_score(X, labels)
        print(f"Silhouette Score: {score:.3f}")
    else:
        print("Cannot compute silhouette score for 1 cluster")

    # Plot clusters
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', alpha=0.5)
    plt.title('PCA Visualization with Clusters')

    plt.subplot(1, 2, 2)
    plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='viridis', alpha=0.5)
    plt.title('t-SNE Visualization with Clusters')

    plt.tight_layout()
    plt.show()

# %% [markdown]
# ### 5.1 K-Means Clustering

# %%
# K-Means Clustering
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

print("K-Means Clustering Results:")
evaluate_clustering(X_scaled, kmeans_labels)

# Add cluster labels to dataframe
df['KMeans_Cluster'] = kmeans_labels

# %% [markdown]
# ### 5.2 Hierarchical (Agglomerative) Clustering

# %%
# Hierarchical Clustering
agg = AgglomerativeClustering(n_clusters=4)
agg_labels = agg.fit_predict(X_scaled)

print("Hierarchical Clustering Results:")
evaluate_clustering(X_scaled, agg_labels)

# Add cluster labels to dataframe
df['Hierarchical_Cluster'] = agg_labels

# %% [markdown]
# ### 5.3 DBSCAN Clustering

# %%
# DBSCAN Clustering
dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

print("DBSCAN Clustering Results:")
print(f"Number of clusters found: {len(np.unique(dbscan_labels))}")
print(f"Number of outliers: {sum(dbscan_labels == -1)}")
evaluate_clustering(X_scaled, dbscan_labels)

# Add cluster labels to dataframe
df['DBSCAN_Cluster'] = dbscan_labels

# %% [markdown]
# ### 5.4 Spectral Clustering

# %%
# Spectral Clustering
spectral = SpectralClustering(n_clusters=4, random_state=42)
spectral_labels = spectral.fit_predict(X_scaled)

print("Spectral Clustering Results:")
evaluate_clustering(X_scaled, spectral_labels)

# Add cluster labels to dataframe
df['Spectral_Cluster'] = spectral_labels

# %% [markdown]
# ## 6. Analyze and Compare Clusters

# %%
# Function to analyze cluster characteristics
def analyze_clusters(df, cluster_col):
    cluster_stats = df.groupby(cluster_col).mean()
    cluster_sizes = df[cluster_col].value_counts().sort_index()

    print(f"\nCluster Sizes for {cluster_col}:")
    print(cluster_sizes)

    print(f"\nCluster Statistics for {cluster_col}:")
    display(cluster_stats)

    # Plot cluster characteristics
    plt.figure(figsize=(15, 8))
    for i, feature in enumerate(features, 1):
        plt.subplot(2, 4, i)
        sns.boxplot(x=cluster_col, y=feature, data=df)
        plt.title(feature)
    plt.tight_layout()
    plt.show()

# %%
# Analyze K-Means clusters
analyze_clusters(df, 'KMeans_Cluster')

# %%
# Analyze Hierarchical clusters
analyze_clusters(df, 'Hierarchical_Cluster')

# %%
# Analyze DBSCAN clusters (excluding outliers)
analyze_clusters(df[df['DBSCAN_Cluster'] != -1], 'DBSCAN_Cluster')

# %%
# Analyze Spectral clusters
analyze_clusters(df, 'Spectral_Cluster')

# %% [markdown]
# ## 7. Save Results

# %%
# Save the dataframe with cluster assignments to a new CSV file
df.to_csv('credit_card_customers_with_clusters.csv', index=False)
print("Results saved to 'credit_card_customers_with_clusters.csv'")

# %% [markdown]
# ## 8. Conclusion
#
# In this notebook, we:
# 1. Generated a sample credit card customer dataset
# 2. Preprocessed the data and performed feature scaling
# 3. Applied 4 different clustering methods:
#    - K-Means
#    - Hierarchical (Agglomerative) Clustering
#    - DBSCAN
#    - Spectral Clustering
# 4. Evaluated and visualized each clustering result
# 5. Analyzed the characteristics of each cluster
# 6. Saved the results with cluster assignments
#
# The best performing method can be selected based on the silhouette score and business interpretability of the clusters.
```
