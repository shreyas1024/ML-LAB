```python
y = "play"
```

```python
columns = list(df.columns)
columns.remove(y)
```

```python
columns
```

```python
prob = {}

prob_y = len(df[df[y] == "Yes"]) / len(df)
prob_n = len(df[df[y] == "No"]) / len(df)

for column in columns:
    prob[column] = {}
    for value in df[column].unique():
        prob[column][value] = {}  # This line is added
        # Use & operator for combined conditions
        prob[column][value]["yes"] = len(df[(df[column] == value) & (df[y] == "Yes")]) / len(df[df[y] == "Yes"] )
        prob[column][value]["no"] = len(df[(df[column] == value) & (df[y] == "No")]) / len(df[df[y] == "No"] )

print("probabilty yes : ", prob_y, "probabbility no : ", prob_n)

for key, value in prob.items():
    print(key)
    for k, v in value.items():
        print(f"\t{k}: {v}")
```

```python
import pandas as pd

def naive_bayes_predict(outlook, temp, humidity, wind):

    # Calculate probabilities for "yes" and "no"
    prob_yes_given_data = prob_y
    for column, value in zip(columns, [outlook, temp, humidity, wind]):
        prob_yes_given_data *= prob[column][value]["yes"]


    prob_no_given_data = prob_n
    for column, value in zip(columns, [outlook, temp, humidity, wind]):
        prob_no_given_data *= prob[column][value]["no"]

    # Normalize the probabilities
    total_prob = prob_yes_given_data + prob_no_given_data

    if total_prob == 0: #Handle the case when total probability is zero.
      return "Cannot determine outcome: total probability is zero."

    normalized_yes_prob = prob_yes_given_data / total_prob
    normalized_no_prob = prob_no_given_data / total_prob


    # Return the prediction
    if normalized_yes_prob > normalized_no_prob:
        return "Yes", normalized_yes_prob
    else:
        return "No", normalized_no_prob


# Example usage:
outlook = "Sunny"
temp = "Cool"
humidity = "High"
wind = "Strong"

prediction, probability = naive_bayes_predict(outlook, temp, humidity, wind)
print(f"Prediction: {prediction}\nProbability of Yes: {probability}, Probability of No: {1 - probability}")
```
