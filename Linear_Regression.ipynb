```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([5.0, 5.1, 5.2, 5.3, 5.4, 5.5, 5.6, 5.7, 5.8])
y = np.array([53, 55, 59, 61, 65, 68, 70, 74, 76])

x_mean = np.mean(x)
y_mean = np.mean(y)

num = np.sum((x - x_mean) * (y - y_mean))
den = np.sum((x - x_mean)**2)
w1 = num / den
w0 = y_mean - (w1 * x_mean)

print("Regression Coefficient (w1):", w1)
print("Intercept (w0):", w0)

mae = np.mean(np.abs(y - (w0 + w1 * x)))
print("Mean Absolute Error (MAE):", mae)

mse = np.mean((y - (w0 + w1 * x))**2)
print("Mean Squared Error (MSE):", mse)

height = 5.9
predicted_weight = w0 + (w1 * height)
print("Predicted weight for height 5'9\":", predicted_weight)

print(f"Regression Line Equation: y = {w0} + {w1}x")

x_line = np.linspace(np.min(x), np.max(x), 100)
y_line = w0 + (w1 * x_line)
plt.plot(x_line, y_line, color='red', label='Regression Line')
plt.scatter(x, y, color='blue', label='Data Points')
plt.scatter(height, predicted_weight, color='green', label='Predicted Point')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
plt.title('Height vs. Weight Regression')
plt.legend()
plt.grid(True)
plt.show()

```
