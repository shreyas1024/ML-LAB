```python
import numpy as np
import matplotlib.pyplot as plt

# Define the step function
def step_function(x):
    return np.where(x > 0, 1, 0)

# Define the perceptron model
def perceptron_model(X, w, b):
    linear_output = np.dot(X, w) + b
    return step_function(linear_output)

# Train the perceptron model
X_train = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_train = np.array([0, 1, 1, 1])  # OR gate labels

# Initialize weights and bias randomly
w = np.random.rand(2)
b = np.random.rand()

# Train the perceptron
for _ in range(10):
    for x, y in zip(X_train, y_train):
        linear_output = np.dot(x, w) + b
        prediction = step_function(linear_output)
        w += 0.1 * (y - prediction) * x
        b += 0.1 * (y - prediction)

# Plot the data points
for i, point in enumerate(X_train):
    if y_train[i] == 0:
        plt.scatter(point[0], point[1], color='red', marker='o', label="class 0" if i == 0 else "")
    else:
        plt.scatter(point[0], point[1], color='blue', marker='x', label="class 1" if i == 1 else "")

# Plot decision boundary
x_vals = np.linspace(-0.1, 1.1, 100)
y_vals = -(w[0] * x_vals + b) / w[1]  # Equation of decision boundary
plt.plot(x_vals, y_vals, label="Decision Boundary", color='green')

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron Decision Boundary for OR Gate")
plt.legend()
plt.grid(True)
plt.show()
```
