```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate dataset with blobs
blob_centers = np.array([
    [0.2, 2.3],
    [-1.5, 2.3],
    [-2.8, 1.8],
    [-2.8, 2.8],
    [-2.8, 1.3]
])
blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1])

X, y = make_blobs(n_samples=2000, centers=blob_centers, cluster_std=blob_std, random_state=7)

# Implement K-Means with educational settings
k = 5  # Number of clusters
# Change the algorithm parameter to either 'elkan' or 'lloyd'
kmeans = KMeans(n_clusters=k, init="random", n_init=1, algorithm="elkan", random_state=7)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot the clusters
def plot_clusters(X, labels, centroids):
    plt.scatter(X[:, 0], X[:, 1], c=labels, s=10, cmap='viridis', alpha=0.6)
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title("K-Means Clustering")
    plt.legend()
    plt.show()

plot_clusters(X, labels, centroids)
```
